# [과제] Seq2Seq 모델을 활용한 불규칙파 예측 및 하이퍼파라미터 최적화

### **1. 과제 개요**
본 과제는 복잡한 시계열 데이터인 '불규칙파'를 Seq2Seq 모델로 예측하며, 모델의 성능 한계를 파악하고 이를 극복하기 위한 최적화 기법(HPO, Architectures, Training Strategies)을 실습하는 것을 목적으로 합니다.

---

### **2. 단계별 과제 내용**

#### **Step 1: 데이터 복잡도에 따른 RNN의 한계 분석**
* **미션:** `nuber_of_waves`를 1, 2, 3으로 증가시키며 불규칙파를 생성하고, 기본 RNN 모델로 학습을 진행하십시오.
* **분석 포인트:** * 파동의 중첩 횟수가 늘어남에 따라 Loss의 수렴 속도와 예측 결과(Scatter plot)가 어떻게 변하는지 기록하십시오.
    * 어느 시점부터 모델이 패턴을 학습하지 못하고 평균값으로 수렴하는지(학습 실패 지점)를 찾고 그 이유를 분석하십시오.

#### **Step 2: Optuna를 활용한 하이퍼파라미터 최적화 (HPO)**
* **미션:** Step 1에서 실패했던 `nuber_of_waves = 3` 문제를 해결하기 위해 **Optuna** 라이브러리를 도입하십시오.
* **최적화 대상:** `hidden_size`, `learning_rate`, `num_layers`, `optimizer` 등.
* **분석 포인트:** 기본 설정값 대비 HPO를 통한 성능 개선 정도를 MSE 지표로 비교하십시오.

#### **Step 3: Long-term Dependency와 시퀀스 길이의 관계**
* **미션:** `input_len`을 기존 30에서 60, 100으로 늘려가며 학습 성능을 비교하십시오.
* **분석 포인트:** * 시퀀스가 길어질수록 Gradient Vanishing 문제가 발생하는지 Loss 추이를 통해 관찰하십시오.
    * 입력 정보가 많아지는 것이 예측에 항상 유리한지, 아니면 RNN의 한계로 인해 성능이 저하되는지 토론하십시오.

#### **Step 4: Recurrent Unit(LSTM, GRU) 성능 비교**
* **미션:** Step 2의 최적화된 조건에서 `nn.RNN`을 `nn.LSTM`과 `nn.GRU`로 교체하여 실험을 재현하십시오.
* **분석 포인트:** * Vanilla RNN에서 풀리지 않던 문제가 LSTM/GRU에서 해결되는지 확인하십시오.
    * 게이트(Gate) 구조가 불규칙파의 장기 의존성 학습에 미치는 영향에 대해 기술하십시오.

#### **Step 5: Teacher Forcing Ratio의 영향 분석**
* **미션:** 학습 시 `teacher_forcing_ratio`를 1.0(Always)에서 0.5, 0.0(Never)으로 변경하며 실험하십시오.
* **분석 포인트:** * 학습 과정에서의 안정성과 실제 테스트(Inference) 단계에서의 누적 오차(Exposure Bias) 발생 여부를 비교 분석하십시오.
